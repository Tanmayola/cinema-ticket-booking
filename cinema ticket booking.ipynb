```python
import pandas as pd

def movie_billing(show_time, movie_name, theater):

  if theater == "AMC":
    billing_amount = 100
  elif theater == "Regal":
    billing_amount = 120
  else:
    billing_amount = 150

  if show_time == "Morning":
    billing_amount -= 2
  elif show_time == "Evening":
    billing_amount += 2

  if movie_name == "The Batman":
    billing_amount += 5

  gst = 0.18
  gst_amount = billing_amount * gst
  mrp = billing_amount + gst_amount

  return mrp


def movie_scheduling(movie_name, theater):

  if theater == "AMC":
    show_time = "Morning"
  elif theater == "Regal":
    show_time = "Evening"
  else:
    show_time = "No show"

  return show_time


def main():
  show_time = input("Enter the show time: ")
  movie_name = input("Enter the movie name: ")
  theater = input("Enter the theater name: ")

  billing_amount = movie_billing(show_time, movie_name, theater)
  gst_amount = billing_amount * 0.18
  mrp = billing_amount + gst_amount

  movies = pd.DataFrame({
    "Movie Name": ["The Batman", "Top Gun: Maverick", "Jurassic World Dominion"],
    "Theater": ["AMC", "Regal", "AMC"],
    "Show Time": ["Morning", "Evening", "Morning"],
    "Base Price": [100, 120, 100],
    "GST": [0.18, 0.18, 0.18],
    "MRP": [11.78, 13.96, 11.78]
  })

  print("---------------------------------------")
  print("Movie Ticket Bill")
  print("Movie Name:", movie_name)
  print("Theater:", theater)
  print("Show Time:", show_time)
  print("Base Price:", billing_amount)
  print("GST:", gst_amount)
  print("MRP:", mrp)
  print("---------------------------------------")
  print("-----------thanks for visiting---------")
  print(movies)


if __name__ == "__main__":
  main()
```
